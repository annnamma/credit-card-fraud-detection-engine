```python
import pandas as pd
import numpy as np

np.random.seed(42)

n_records = 20000

data = pd.DataFrame({
    'user_id': np.random.randint(1000,1050,n_records),
    'amount': np.random.exponential(scale=45,size=n_records),
    'hours_since_last_tx': np.random.exponential(scale=18,size=n_records),
    'distance_from_home_miles': np.random.exponential(scale=8,size=n_records),
    'device_type': np.random.choice(
        ['Mobile','Desktop','Tablet'],
        n_records,
        p=[0.7,0.2,0.1]
    ),
    'is_fraud':0
})

data.head()
```

```python
fraud_indices = np.random.choice(
    n_records,
    size=200,
    replace=False
)

data.loc[fraud_indices,'amount'] = np.random.uniform(
    800,
    4500,
    size=200
)

data.loc[fraud_indices,'hours_since_last_tx'] = np.random.uniform(
    0.01,
    0.2,
    size=200
)

data.loc[fraud_indices,'distance_from_home_miles'] = np.random.uniform(
    300,
    4000,
    size=200
)

data.loc[fraud_indices,'device_type'] = 'Unknown'

data.loc[fraud_indices,'is_fraud'] = 1
```
